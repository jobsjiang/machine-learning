#### 3.IPython魔术命令
* %Ismagic:找出全部命令
* %debug:交互式debug
* %store:在notebook之间传递变量
  1. %store [variable] 存储变量
  2. %store -r [variable] 读取/检索存储变量
* %who:列出所有全局变量
* %%time:计时魔法命令
* %%writefile:向文件写入单元格内容
