```python
# 1.2.1 Ordinal Encoding åºæ•°ç¼–ç  åºæ•°ç¼–ç å°†ç±»åˆ«å˜é‡è½¬åŒ–ä¸ºä¸€åˆ—åºæ•°å˜é‡ï¼ŒåŒ…å«ä»1åˆ°ç±»åˆ«æ•°é‡çš„æ•´æ•°
import numpy as np
import pandas as pd
from category_encoders import OrdinalEncoder
```

```python
# éšæœºç”Ÿæˆä¸€äº›è®­ç»ƒé›†
train_set = pd.DataFrame(np.array([['male',10],['female', 20], ['male',10], 
                       ['female',20],['female',15]]),
             columns = ['Sex','Type'])
train_y = np.array([False, True, True, False, False])
# éšæœºç”Ÿæˆä¸€äº›æµ‹è¯•é›†, å¹¶æœ‰æ„è®©å…¶åŒ…å«æœªåœ¨è®­ç»ƒé›†å‡ºç°è¿‡çš„ç±»åˆ«ä¸ç¼ºå¤±å€¼
test_set = pd.DataFrame(np.array([['female',20],['male', 20], ['others',15], 
                       ['male',20],['female',40], ['male', 25]]),
             columns = ['Sex','Type'])
test_set.loc[4,'Type'] = np.nan
```

```python
encoder = OrdinalEncoder(cols=['Sex','Type'],handle_unknown='value',handle_missing='value').fit(train_set,train_y)
# å°† handle_unknownè®¾ä¸ºâ€˜valueâ€™ï¼Œå³æµ‹è¯•é›†ä¸­çš„æœªçŸ¥ç‰¹å¾å€¼å°†è¢«æ ‡è®°ä¸º-1
# å°† handle_missingè®¾ä¸ºâ€˜valueâ€™ï¼Œå³æµ‹è¯•é›†ä¸­çš„ç¼ºå¤±å€¼å°†è¢«æ ‡è®°ä¸º-2
# å…¶ä»–çš„é€‰æ‹©ä¸ºï¼šâ€˜errorâ€™ï¼šå³æŠ¥é”™ï¼›â€˜return_nanâ€™ï¼šå³æœªçŸ¥å€¼/ç¼ºå¤±ä¹‹è¢«æ ‡è®°ä¸ºnan 
encoded_train = encoder.transform(train_set)
encoded_test = encoder.transform(test_set)
```

```python
# 1.2.3 Hashing Encoding å“ˆå¸Œç¼–ç 
'''
å“ˆå¸Œç¼–ç åŸºäºç‰¹å¾å“ˆå¸Œçš„æ–¹æ³•ã€‚å®ƒå°†å“ˆå¸Œå‡½æ•°åº”ç”¨äºå˜é‡ï¼Œå°†ä»»æ„æ•°é‡çš„å˜é‡ä»¥ä¸€å®šçš„è§„åˆ™æ˜ å°„åˆ°ç»™å®šæ•°é‡çš„å˜é‡ã€‚
ç‰¹å¾å“ˆå¸Œå¯èƒ½ä¼šå¯¼è‡´è¦ç´ ä¹‹é—´å‘ç”Ÿå†²çªã€‚ä½†å“ˆå¸Œç¼–ç çš„ä¼˜ç‚¹æ˜¯å®ƒä¸éœ€è¦åˆ¶å®šå’Œç»´æŠ¤åŸå˜é‡ä¸æ–°å˜é‡ä¹‹é—´çš„æ˜ å°„å…³ç³»ã€‚
å› æ­¤ï¼Œå“ˆå¸Œç¼–ç å™¨çš„å¤§å°åŠå¤æ‚ç¨‹åº¦ä¸éšæ•°æ®ç±»åˆ«çš„å¢å¤šè€Œå¢å¤šã€‚
'''
import numpy as np
import pandas as pd
from category_encoders.hashing import HashingEncoder
# category_encoders ç›´æ¥æ”¯æŒdataframe

# éšæœºç”Ÿæˆä¸€äº›è®­ç»ƒé›†
train_set = pd.DataFrame(np.array([['male',10],['female', 20], ['male',10], 
                       ['female',20],['female',15]]),
             columns = ['Sex','Type'])
train_y = np.array([False, True, True, False, False])

# éšæœºç”Ÿæˆä¸€äº›æµ‹è¯•é›†, å¹¶æœ‰æ„è®©å…¶åŒ…å«æœªåœ¨è®­ç»ƒé›†å‡ºç°è¿‡çš„ç±»åˆ«ä¸ç¼ºå¤±å€¼
test_set = pd.DataFrame(np.array([['female',20],['male', 20], ['others',15], 
                       ['male',20],['female',40], ['male', 25]]),
             columns = ['Sex','Type'])
test_set.loc[4,'Type'] = np.nan
```

```python
encoder = HashingEncoder(cols=['Sex','Type'],n_components=5).fit(train_set,train_y)
encoded_train = encoder.transform(train_set) # è½¬æ¢è®­ç»ƒé›†
encoded_test = encoder.transform(test_set) # è½¬æ¢æµ‹è¯•é›†
# å°†ä¸¤åˆ—çš„æ•°æ®é›†å“ˆå¸Œç¼–ç ä¸º5åˆ—
# å“ˆå¸Œç¼–ç ç»“æœä¸è®­ç»ƒé›†/æµ‹è¯•é›†ä¸­çš„å†…å®¹æ— å…³
# åªè¦åˆ—ååŒ¹é…ï¼Œæˆ‘ä»¬å°±å¯ä»¥åœ¨ä»»ä½•æ–°æ•°æ®é›†ä¸Šä½¿ç”¨å“ˆå¸Œç¼–ç æ–¹æ³•
# ç¼–ç ç»“æœä»…ç”±å“ˆå¸Œå‡½æ•°ç¡®å®š

# é€šå¸¸å“ˆå¸Œç¼–ç åº”ç”¨äºæ›´é«˜å’Œæ›´ç¨€ç–çš„ç»´ç©ºé—´ï¼Œè¿™é‡Œä»¥ä¸¤ä¸ªå˜é‡ä½œä¸ºå“ˆå¸Œç¼–ç çš„ä¾‹å­
# ä»¥æµ‹è¯•é›†ç»“æœä¸ºä¾‹
```

```python
# 1.2.4 Helmert Encoding Helmertç¼–ç 
'''
Helmertç¼–ç é€šå¸¸åœ¨è®¡é‡ç»æµå­¦ä¸­ä½¿ç”¨ã€‚åœ¨Helmertç¼–ç ï¼ˆåˆ†ç±»ç‰¹å¾ä¸­çš„æ¯ä¸ªå€¼å¯¹åº”äºHelmertçŸ©é˜µä¸­çš„ä¸€è¡Œï¼‰ä¹‹åï¼Œ
çº¿æ€§æ¨¡å‹ä¸­ç¼–ç åçš„å˜é‡ç³»æ•°å¯ä»¥åæ˜ åœ¨ç»™å®šè¯¥ç±»åˆ«å˜é‡æŸä¸€ç±»åˆ«å€¼çš„æƒ…å½¢ä¸‹å› å˜é‡çš„å¹³å‡å€¼ä¸ç»™å®šè¯¥ç±»
åˆ«å…¶ä»–ç±»åˆ«å€¼çš„æƒ…å½¢ä¸‹å› å˜é‡çš„å¹³å‡å€¼çš„å·®å€¼ã€‚åœ¨category_encodersåŒ…ä¸­å®ç°çš„Helmertç¼–ç ä¸º
åå‘Helmertç¼–ç 
'''
import numpy as np
import pandas as pd
from category_encoders import HelmertEncoder
# category_encoders ç›´æ¥æ”¯æŒdataframe

# éšæœºç”Ÿæˆä¸€äº›è®­ç»ƒé›†
train_set = pd.DataFrame(np.array([['male',10],['female', 20], ['male',10], 
                       ['female',20],['female',15]]),
             columns = ['Sex','Type'])
train_y = np.array([False, True, True, False, False])

# éšæœºç”Ÿæˆä¸€äº›æµ‹è¯•é›†, å¹¶æœ‰æ„è®©å…¶åŒ…å«æœªåœ¨è®­ç»ƒé›†å‡ºç°è¿‡çš„ç±»åˆ«ä¸ç¼ºå¤±å€¼
test_set = pd.DataFrame(np.array([['female',20],['male', 20], ['others',15], 
                       ['male',20],['female',40], ['male', 25]]),
             columns = ['Sex','Type'])
test_set.loc[4,'Type'] = np.nan
```

```python
encoder = HelmertEncoder(cols=['Sex', 'Type'],handle_unknown='indicator',handle_missing='indicator').fit(train_set,train_y)
encoded_train = encoder.transform(train_set) # è½¬æ¢è®­ç»ƒé›†
encoded_test = encoder.transform(test_set) # è½¬æ¢æµ‹è¯•é›†
# ä»¥æµ‹è¯•é›†ç»“æœä¸ºä¾‹
encoded_test

# åœ¨Helmertç¼–ç ä¸­ï¼š
  
# å˜é‡ Sex => å˜ä¸ºäº†4ä¸ªæ–°å˜é‡ï¼ˆåŒ…å«å¸¸æ•°é¡¹ï¼‰: 'male' => [ 1. -1. -1. -1.];
#                                      'female' => [ 1.  1. -1. -1.];
#                                      æœªçŸ¥ =>  [ 1.  0.  0.  3.];
#                                      ç¼ºå¤± => [ 1.  0.  2. -1.];
    
# å˜é‡ Type => å˜ä¸ºäº†5ä¸ªæ–°å˜é‡ï¼ˆåŒ…å«å¸¸æ•°é¡¹ï¼‰: 10 => [ 1. -1. -1. -1. -1.];
#                                        20 => [ 1.  1. -1. -1. -1.];, 
#                                        15 => [ 1.  0.  2. -1. -1.];
#                                        æœªçŸ¥ =>  [ 1.  0.  0.  0.  4.];
#                                        ç¼ºå¤± => [ 1.  0.  0.  3. -1.];
```

```python
# 1.2.5 Sum(Deviation)Encoding åå·®ç¼–ç 
'''
åå·®ç¼–ç ä¹Ÿé€šå¸¸åœ¨è®¡é‡ç»æµå­¦ä¸­è¢«ä½¿ç”¨ã€‚åå·®ç¼–ç åï¼Œ
çº¿æ€§æ¨¡å‹çš„ç³»æ•°å¯ä»¥åæ˜ è¯¥ç»™å®šè¯¥ç±»åˆ«å˜é‡å€¼çš„æƒ…å†µä¸‹å› å˜é‡çš„å¹³å‡å€¼ä¸å…¨å±€å› å˜é‡çš„å¹³å‡å€¼çš„å·®å¼‚
'''
import numpy as np
import pandas as pd
from category_encoders.sum_coding import SumEncoder
# category_encoders ç›´æ¥æ”¯æŒdataframe

# éšæœºç”Ÿæˆä¸€äº›è®­ç»ƒé›†
train_set = pd.DataFrame(np.array([['male',10],['female', 20], ['male',10], 
                       ['female',20],['female',15]]),
             columns = ['Sex','Type'])
train_y = np.array([False, True, True, False, False])

# éšæœºç”Ÿæˆä¸€äº›æµ‹è¯•é›†, å¹¶æœ‰æ„è®©å…¶åŒ…å«æœªåœ¨è®­ç»ƒé›†å‡ºç°è¿‡çš„ç±»åˆ«ä¸ç¼ºå¤±å€¼
test_set = pd.DataFrame(np.array([['female',20],['male', 20], ['others',15], 
                       ['male',20],['female',40], ['male', 25]]),
             columns = ['Sex','Type'])
test_set.loc[4,'Type'] = np.nan
```

```python
encoder = SumEncoder(cols=['Sex', 'Type'], 
                     handle_unknown='indicator', 
                     handle_missing='indicator').fit(train_set,train_y) # åœ¨è®­ç»ƒé›†ä¸Šè®­ç»ƒ
encoded_train = encoder.transform(train_set) # è½¬æ¢è®­ç»ƒé›†
encoded_test = encoder.transform(test_set) # è½¬æ¢æµ‹è¯•é›†
# å°† handle_unknownè®¾ä¸ºâ€˜indicatorâ€™ï¼Œå³ä¼šæ–°å¢ä¸€åˆ—æŒ‡ç¤ºæœªçŸ¥ç‰¹å¾å€¼
# å°† handle_missingè®¾ä¸ºâ€˜indicatorâ€™ï¼Œå³ä¼šæ–°å¢ä¸€åˆ—æŒ‡ç¤ºç¼ºå¤±å€¼
# å…¶ä»–çš„handle_unknown/handle_missing çš„é€‰æ‹©ä¸ºï¼š
# â€˜errorâ€™ï¼šå³æŠ¥é”™; â€˜return_nanâ€™ï¼šå³æœªçŸ¥å€¼/ç¼ºå¤±ä¹‹è¢«æ ‡è®°ä¸ºnan; â€˜valueâ€™ï¼šå³æœªçŸ¥å€¼/ç¼ºå¤±ä¹‹è¢«æ ‡è®°ä¸º0

# ä»¥æµ‹è¯•é›†ç»“æœä¸ºä¾‹
encoded_test

# åœ¨Helmertç¼–ç ä¸­ï¼š
  
# å˜é‡ Sex => å˜ä¸ºäº†4ä¸ªæ–°å˜é‡ï¼ˆåŒ…å«å¸¸æ•°é¡¹ï¼‰: 'male' => [ 1.  1.  0.  0.];
#                                      'female' => [ 1.  0.  1.  0.];
#                                       æœªçŸ¥ =>  [ 1. -1. -1. -1.];
#                                       ç¼ºå¤± => [ 1.  0.  0.  1.];
    
# å˜é‡ Type => å˜ä¸ºäº†5ä¸ªæ–°å˜é‡ï¼ˆåŒ…å«å¸¸æ•°é¡¹ï¼‰: 10 => [ 1.  1.  0.  0.  0.];
#                                        20 => [ 1.  0.  1.  0.  0.];, 
#                                        15 => [ 1.  0.  0.  1.  0.];
#                                        æœªçŸ¥ =>  [ 1. -1. -1. -1. -1.];
#                                        ç¼ºå¤± => [ 1.  0.  0.  0.  1.];
```

```python
# å¯ä»¥é€šè¿‡å¦‚ä¸‹ä»£ç è®¡ç®—å˜é‡Typeçš„Deviation çŸ©é˜µ
from patsy.contrasts import Sum
levels = [1,2,3,4,5] # 3ä¸ªå˜é‡å€¼ + 1ä¸ªæœªçŸ¥å€¼ + 1ä¸ªç¼ºå¤±å€¼
contrast = Sum().code_with_intercept(levels)
print(contrast.matrix) # ç¬¬ä¸€åˆ—ä¸ºå¸¸æ•°é¡¹
```

```python
# 1.2.6 TargetEncoding ç›®æ ‡ç¼–ç 
import numpy as np
import pandas as pd
from category_encoders.target_encoder import TargetEncoder
# category_encoders ç›´æ¥æ”¯æŒdataframe

# éšæœºç”Ÿæˆä¸€äº›è®­ç»ƒé›†
train_set = pd.DataFrame(np.array([['male',10],['female', 20], ['male',10], 
                       ['female',20],['female',15]]),
             columns = ['Sex','Type'])
train_y = np.array([False, True, True, False, False])

# éšæœºç”Ÿæˆä¸€äº›æµ‹è¯•é›†, å¹¶æœ‰æ„è®©å…¶åŒ…å«æœªåœ¨è®­ç»ƒé›†å‡ºç°è¿‡çš„ç±»åˆ«ä¸ç¼ºå¤±å€¼
test_set = pd.DataFrame(np.array([['female',20],['male', 20], ['others',15], 
                       ['male',20],['female',40], ['male', 25]]),
             columns = ['Sex','Type'])
test_set.loc[4,'Type'] = np.nan
```

```python
encoder = TargetEncoder(cols=['Sex','Type'], 
                        handle_unknown='value',  
                        handle_missing='value').fit(train_set,train_y) # åœ¨è®­ç»ƒé›†ä¸Šè®­ç»ƒ
encoded_train = encoder.transform(train_set) # è½¬æ¢è®­ç»ƒé›†
encoded_test = encoder.transform(test_set) # è½¬æ¢æµ‹è¯•é›†

# handle_unknown å’Œ handle_missing è¢«è®¾å®šä¸º 'value'
# åœ¨ç›®æ ‡ç¼–ç ä¸­ï¼Œhandle_unknown å’Œ handle_missing ä»…æ¥å— â€˜errorâ€™, â€˜return_nanâ€™ åŠ â€˜valueâ€™ è®¾å®š
# ä¸¤è€…çš„é»˜è®¤å€¼å‡ä¸º â€˜valueâ€™, å³å¯¹æœªçŸ¥ç±»åˆ«æˆ–ç¼ºå¤±å€¼å¡«å……è®­ç»ƒé›†çš„å› å˜é‡å¹³å‡å€¼

encoded_test # ç¼–ç åçš„å˜é‡æ•°ä¸åŸç±»åˆ«å˜é‡æ•°ä¸€è‡´
```
