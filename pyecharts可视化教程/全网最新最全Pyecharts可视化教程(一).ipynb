```python
from pyecharts.charts import Bar
from pyecharts import options as opts
```

```python
from pyecharts.globals import CurrentConfig
CurrentConfig.ONLINE_HOST = "https://cdn.kesci.com/lib/pyecharts_assets/"
```

```python
# 示例数据
cate = ['Apple','Huawei','Xiaomi','Oppo','Samsung']
data1 = [123,153,89,107,98]
data2 = [56,77,93,68,45]
# 1.x版本支持链式调用
bar = (Bar()
       .add_xaxis(cate) # 设置x轴的数据
       .add_yaxis('线上渠道',data1)
       .add_yaxis('线下渠道',data2)
       .set_global_opts(title_opts=opts.TitleOpts(title='Bar-基本示例',subtitle='我是副标题'),
                        toolbox_opts=opts.ToolboxOpts(), # 可以理解为是工具栏
                        legend_opts=opts.LegendOpts(is_show=True) # 图例要不要展示出来
                       )
)
```

```python
# 在jupyter notebook总渲染
bar.render_notebook()
```
